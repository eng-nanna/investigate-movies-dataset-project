### Data Cleaning

#### First of all make a copy of the dataframe to use in the cleaning step

```python
df_movies_clean = df_movies.copy()
```

#### change the data type of 'release_date' column to be date instead of being string

```python
def modify_release_date(row):
    release_date_entry = '/'.join(row['release_date'].split('/')[0:2]) + '/' + str(row['release_year'])
    return release_date_entry
df_movies_clean['release_date'] = df_movies_clean.apply(modify_release_date, axis=1)
```

```python
df_movies_clean.tail(25)
```

#### replace the zero values in some columns with the mean of the column

```python
df_movies_clean['budget']=df_movies_clean['budget'].mask(df_movies_clean['budget']==0).fillna(df_movies_clean['budget'].mean())
```

```python
df_movies_clean['budget_adj']=df_movies_clean['budget_adj'].mask(df_movies_clean['budget_adj']==0).fillna(df_movies_clean['budget_adj'].mean())
```

```python
df_movies_clean['revenue']=df_movies_clean['revenue'].mask(df_movies_clean['revenue']==0).fillna(df_movies_clean['revenue'].mean())
```

```python
df_movies_clean['revenue_adj']=df_movies_clean['revenue_adj'].mask(df_movies_clean['revenue_adj']==0).fillna(df_movies_clean['revenue_adj'].mean())
```

```python
df_movies_clean['runtime']=df_movies_clean['runtime'].mask(df_movies_clean['revenue']==0).fillna(df_movies_clean['runtime'].mean())
```

```python
df_movies_clean.head(25)
```

Drop columns that won't be helpful in analyzing the data set and has majority of zero values

```python
df_movies_clean.drop(columns=['homepage', 'tagline', 'keywords', 'overview'], inplace=True)
df_movies_clean.info()
```

Drop rows with NaN values

```python
df_movies_clean = df_movies_clean[df_movies_clean['imdb_id'].notna()]
df_movies_clean = df_movies_clean[df_movies_clean['director'].notna()]
df_movies_clean = df_movies_clean[df_movies_clean['genres'].notna()]
df_movies_clean = df_movies_clean[df_movies_clean['cast'].notna()]
df_movies_clean = df_movies_clean[df_movies_clean['production_companies'].notna()]
df_movies_clean.info()
```

```python
genre = ['Drama', 'Romance', 'Action', 'Horror', 'Comedy', 'Adventure', 'Mystery', 'Thriller', 'Science Fiction', 'Western', 
         'Fantasy', 'Crime','Family', 'Animation', 'War', 'History']

for i in genre:
    def matcher(x):
        if i.lower() in x.lower():
            return i
        else:
             return np.nan
    df_movies_clean['genre_'+i] = df_movies_clean['genres'].apply(matcher)
    
df_movies_clean.drop(columns=['genres'], inplace=True)
df_movies_clean.info()
```

### Research Question 1: Which genres has hightest counts from year to year

```python
# Use this, and more code cells, to explore your data. Don't forget to add
#   Markdown cells to document your observations and findings.
fig, ax = plt.subplots(figsize=(18,14))
ax.hist(df_movies_clean.groupby(['release_year']).count()['genre_Drama'], bins=60, alpha=1.0, label='Drama')
ax.hist(df_movies_clean.groupby(['release_year']).count()['genre_Romance'], bins=60, alpha=0.8, label='Romance')
ax.hist(df_movies_clean.groupby(['release_year']).count()['genre_Comedy'], bins=60, alpha=0.6, label='Comedy')
ax.hist(df_movies_clean.groupby(['release_year']).count()['genre_Action'], bins=60, alpha=0.4, label='Action')
ax.hist(df_movies_clean.groupby(['release_year']).count()['genre_Thriller'], bins=60, alpha=0.2, label='Thriller')
ax.set_title('Count of Drama, Romance, Comedy, Action and Thriller movies Yearly')

#ax.set_xticks(df_movies_clean.release_year.values)
#ax.set_xticklabels(df_movies_clean.index)
ax.set_xlabel('Year')
ax.set_ylabel('Count')
ax.legend(loc='upper right')
plt.show()
```

The visual shows that Drama movies has the highest count over the years, followed by the Comedy movies.
anoher point is that Romance movies had highest count in earlier years

### Research Question 2  (Which genres has hightest revenue from year to year

```python
fig, ax = plt.subplots(figsize=(18,14))
ax.hist(df_movies_clean.groupby(['release_year','genre_Drama'])['revenue'].sum(), bins=50, alpha=1.0, label='Drama')
ax.hist(df_movies_clean.groupby(['release_year','genre_Romance'])['revenue'].sum(), bins=50, alpha=0.8, label='Romance')
ax.hist(df_movies_clean.groupby(['release_year','genre_Comedy'])['revenue'].sum(), bins=50, alpha=0.6, label='Comedy')
ax.hist(df_movies_clean.groupby(['release_year','genre_Action'])['revenue'].sum(), bins=50, alpha=0.4, label='Action')
ax.hist(df_movies_clean.groupby(['release_year','genre_Thriller'])['revenue'].sum(), bins=50, alpha=0.2, label='Thriller')

ax.set_title('Yearly revenues of Drama, Romance, Comedy, Action and Thriller movies')

#ax.set_xticks(df_movies_clean.release_year.values)
#ax.set_xticklabels(df_movies_clean.index, rotation=35)
ax.set_xlabel('Year')
ax.set_ylabel('Revenue')
ax.legend(loc='upper right')
plt.show()
```

The visual shows that Action movies has the highest revenues in the early years, followed by the Comedy movies.
